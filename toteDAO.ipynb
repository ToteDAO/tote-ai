```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
count_vectorizer =  CountVectorizer(stop_words= stopwords, max_features= 1000)
```

```python
from nltk.stem.snowball import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
analyzer = CountVectorizer().build_analyzer()
```

```python
def stemmed_words(doc):
  stemmed_words = [stemmer.stem(w) for w in analyzer(doc)]
  non_stop_words = [word for word in stemmed_words if not stopwords]
  return non_stop_words
```

```python
count_vectorizer =  CountVectorizer(stop_words= stopwords, max_features= 1000)
feature_vector =  count_vectorizer.fit(appearance.json)
train_ds_features =  count_vectorizer.transform(appearance.Text)
features = feature_vector.get_feature_names()
features_counts = np.sum(train_ds_features.toarray(), axis = 0)
features_counts = pd.DataFrame(dict(features =  features, counts = features_counts))
features_counts.sort_values("counts", ascending= False)[0:15]
```
